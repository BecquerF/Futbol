Además contemplaremos los periodos dónde se han convertido y recibido goles, dónde podremos analizar momentos fuertes y débiles en gral. y por categoría, dejando una tendencia de mayor y menor concentración en el juego.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
#Utilizar archivo CSV
df = pd.read_csv('Procedencia Goles 2024 Apertura-Intermedio.csv', sep=';')
df.head()
```

```python
#SELECCIONAR COLUMNAS PARA LUEGO PROMEDIAR

df_danubio = df.drop(columns=['FECHA','RIVAL','TORNEO']) # Remove the extra brackets around the column names
df_danubio_suma = df_danubio.groupby('CATEGORIA').sum()
df_danubio_suma.head()
```

```python
#DF de df_danubio seleccionando columnas de Tiempo
df_danubio_tiempo_categoria = df_danubio.groupby('CATEGORIA').sum()

df_danubio_tiempo_categoria = df_danubio_tiempo_categoria.reset_index()
df_danubio_tiempo = df_danubio_tiempo_categoria[['CATEGORIA','0-15 AF','15-30 AF','30-45 AF','45-60 AF','60-75 AF','75-90 AF','0-15 EC','15-30 EC','30-45 EC','45-60 EC','60-75 EC','75-90 EC']]
df_danubio_tiempo.head()
```

```python
# Calculate the sum of numeric columns only, excluding 'CATEGORIA'
df_danubio_tiempo_total = df_danubio_tiempo.drop(columns=['CATEGORIA']).sum()

# Convert the result to a DataFrame
df_danubio_tiempo_total = pd.DataFrame(df_danubio_tiempo_total)

# Select only the numeric columns for division
numeric_columns = df_danubio_tiempo_total.select_dtypes(include=np.number).columns
df_danubio_tiempo_total[numeric_columns] = df_danubio_tiempo_total[numeric_columns]/ 5

df_danubio_tiempo_total = df_danubio_tiempo_total.transpose()
df_danubio_tiempo_total

```

```python
#Df de la Categoría U14 goles y tiempo
df_u14 = df_danubio_tiempo[df_danubio_tiempo['CATEGORIA'] == 'U14']
#Sumar valores
df_u14 = df_u14.sum()
df_u14 = df_u14.drop(['CATEGORIA'])
df_u14 = pd.DataFrame(df_u14)
df_u14 = df_u14.transpose()

df_u14
```

```python
#Df de la Categoría U15 goles y tiempo
df_u15 = df_danubio_tiempo[df_danubio_tiempo['CATEGORIA'] == 'U15']
#Sumar valores
df_u15 = df_u15.sum()
df_u15 = df_u15.drop(['CATEGORIA'])
df_u15 = pd.DataFrame(df_u15)
df_u15 = df_u15.transpose()

df_u15
```

```python
#Df de la Categoría U16 goles y tiempo
df_u16 = df_danubio_tiempo[df_danubio_tiempo['CATEGORIA'] == 'U16']
#Sumar valores
df_u16 = df_u16.sum()
df_u16 = df_u16.drop(['CATEGORIA'])
df_u16 = pd.DataFrame(df_u16)
df_u16 = df_u16.transpose()

df_u16
```

```python
#Df de la Categoría U17 goles y tiempo
df_u17 = df_danubio_tiempo[df_danubio_tiempo['CATEGORIA'] == 'U17']
#Sumar valores
df_u17 = df_u17.sum()
df_u17 = df_u17.drop(['CATEGORIA'])
df_u17 = pd.DataFrame(df_u17)
df_u17 = df_u17.transpose()
df_u17
```

```python
#Df de la Categoría U19 goles y tiempo
df_u19 = df_danubio_tiempo[df_danubio_tiempo['CATEGORIA'] == 'U19']
#Sumar valores
df_u19 = df_u19.sum()
df_u19 = df_u19.drop(['CATEGORIA'])
df_u19 = pd.DataFrame(df_u19)
df_u19 = df_u19.transpose()

df_u19
```

```python
#DF de df_danubio seleccionando columnas de Acciones



df_danubio_acciones = df_danubio.groupby(['CATEGORIA', 'JUGADA ELABORADA AF', 'PRESIÓN EN SALIDA AF', 'SAQUE LATERAL AF', 'TIRO LIBRE AF', 'CÓRNERS AF', 'CONTRAGOLPE AF', 'PENAL AF', 'JUGADA ELABORADA  EC', 'PÉRDIDA EN SALIDA EC', 'SAQUE LATERAL EC', 'TIRO LIBRE EC', 'CÓRNERS EC', 'CONTRAGOLPE EC', 'PENAL EC']).sum()
df_danubio_acciones = df_danubio.drop(columns=['0-15 AF', '15-30 AF', '30-45 AF', '45-60 AF', '60-75 AF', '75-90 AF',
           '0-15 EC', '15-30 EC', '30-45 EC', '45-60 EC', '60-75 EC', '75-90 EC'])

df_danubio_acciones
```

```python
df_danubio_acciones_unificado = df_danubio_acciones.groupby('CATEGORIA').sum()
df_danubio_acciones_unificado
```

```python
import plotly.graph_objects as go

# Definir las variantes en el eje X (columnas, excepto la de CATEGORIA)
variants = ['JUGADA ELABORADA AF', 'CONTRAGOLPE AF', 'PRESIÓN EN SALIDA AF', 'SAQUE LATERAL AF', 'TIRO LIBRE AF',
            'CÓRNERS AF',  'PENAL AF']

# Crear el gráfico de barras agrupadas
fig_0 = go.Figure()

# Agregar una barra para cada categoría
# Iterate through the index of the DataFrame which now contains the categories
for category in df_danubio_acciones_unificado.index:
    fig_0.add_trace(go.Bar(
        x=variants,
        # Access data using .loc with the category as the index
        y=df_danubio_acciones_unificado.loc[category, variants].values,
        name=category,
        text=df_danubio_acciones_unificado.loc[category, variants].values,
        textposition='auto',
        marker_color=paleta_colores[category]  # Use the color from paleta_colores
    ))
# Actualizar el diseño del gráfico
fig_0.update_layout(
    title='Comparativa de Procedencia de Goles A Favor por Categoría',
    width=1000, height=1000, title_font=dict(size=24),
    xaxis_title='Procedencia de Goles',
    yaxis_title='Cantidad',
    barmode='group',
    template='plotly_dark',  # Estilo del gráfico
    xaxis_tickangle=-45  # Rotar etiquetas del eje X para mejor legibilidad
)

# Mostrar el gráfico
fig_0.show()
```
